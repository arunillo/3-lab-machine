```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Загрузка данных
path = kagglehub.dataset_download("selfishgene/historical-hourly-weather-data")
df = pd.read_csv(path + '/temperature.csv')

print("Исходные данные:")
print(f"Форма: {df.shape}")
print(df.info())
print("\nПервые 5 строк:")
print(df.head())

# Анализ исходных данных
print("\n" + "="*50)
print("АНАЛИЗ ИСХОДНЫХ ДАННЫХ")
print("="*50)

# Проверяем наличие проблем
has_missing = df.isnull().sum().sum() > 0
has_categorical = len(df.select_dtypes(exclude=[np.number]).columns) > 0

print(f"Пропуски: {'ЕСТЬ' if has_missing else 'НЕТ'}")
print(f"Категориальные данные: {'ЕСТЬ' if has_categorical else 'НЕТ'}")

# 1. Генерация проблем если их нет
print("\n" + "="*50)
print("ГЕНЕРАЦИЯ ПРОБЛЕМ")
print("="*50)

if not has_missing:
    print("Генерация пропусков...")
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols[:3]:  # первые 3 числовые колонки
        mask = np.random.random(len(df)) < 0.15
        df.loc[mask, col] = np.nan
        print(f"Добавлено {mask.sum()} пропусков в {col}")

if not has_categorical:
    print("Генерация категориальных данных...")
    # Создаем категориальную колонку на основе температуры
    if 'Vancouver' in df.columns:
        df['temp_category'] = pd.cut(df['Vancouver'], bins=3, labels=['cold', 'medium', 'hot'])
        print("Добавлена колонка 'temp_category'")

# 2. ВИЗУАЛИЗАЦИЯ ДО ОБРАБОТКИ
print("\n" + "="*50)
print("ВИЗУАЛИЗАЦИЯ ДО ОБРАБОТКИ")
print("="*50)

plt.figure(figsize=(20, 15))

# Пропуски
plt.subplot(3, 3, 1)
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis')
plt.title('Пропущенные значения (до обработки)')

# Распределения первых 4 числовых колонок
numeric_cols = df.select_dtypes(include=[np.number]).columns[:4]
for i, col in enumerate(numeric_cols, 2):
    plt.subplot(3, 3, i)
    if i <= 5:  # только первые 4 графика
        df[col].hist(bins=30, alpha=0.7)
        plt.title(f'Распределение {col}')
        plt.xlabel('Значение')
        plt.ylabel('Частота')

# Boxplot для выбросов
plt.subplot(3, 3, 6)
df[numeric_cols].boxplot()
plt.title('Boxplot (выбросы)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 3. ПРЕОБРАЗОВАНИЕ ДАННЫХ
print("\n" + "="*50)
print("ПРЕОБРАЗОВАНИЕ ДАННЫХ")
print("="*50)

df_processed = df.copy()

# Преобразование категориальных данных в числовые
categorical_cols = df_processed.select_dtypes(exclude=[np.number]).columns
if len(categorical_cols) > 0:
    print("Преобразование категориальных данных...")
    for col in categorical_cols:
        df_processed[col] = pd.factorize(df_processed[col])[0]
        print(f"Колонка '{col}' преобразована в числовой формат")

# Заполнение пропусков
print("Заполнение пропусков...")
numeric_cols = df_processed.select_dtypes(include=[np.number]).columns
df_processed[numeric_cols] = df_processed[numeric_cols].fillna(df_processed[numeric_cols].median())

# Стандартизация
print("Стандартизация данных...")
scaler = StandardScaler()
df_processed[numeric_cols] = scaler.fit_transform(df_processed[numeric_cols])

# 4. ВИЗУАЛИЗАЦИЯ ПОСЛЕ ОБРАБОТКИ
print("\n" + "="*50)
print("ВИЗУАЛИЗАЦИЯ ПОСЛЕ ОБРАБОТКИ")
print("="*50)

plt.figure(figsize=(20, 10))

# Пропуски после обработки
plt.subplot(2, 3, 1)
sns.heatmap(df_processed.isnull(), cbar=True, yticklabels=False, cmap='viridis')
plt.title('Пропуски после обработки')

# Распределения после стандартизации
for i, col in enumerate(numeric_cols[:4], 2):
    plt.subplot(2, 3, i)
    df_processed[col].hist(bins=30, alpha=0.7, color='green')
    plt.title(f'После стандартизации: {col}')
    plt.xlabel('Стандартизированные значения')
    plt.ylabel('Частота')

# Сравнение до/после для первой колонки
if len(numeric_cols) > 0:
    plt.subplot(2, 3, 6)
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    df[numeric_cols[0]].hist(bins=30, alpha=0.7, color='blue')
    plt.title(f'До: {numeric_cols[0]}')
    plt.subplot(1, 2, 2)
    df_processed[numeric_cols[0]].hist(bins=30, alpha=0.7, color='red')
    plt.title(f'После: {numeric_cols[0]}')
    plt.tight_layout()

plt.tight_layout()
plt.show()

# 5. РЕЗУЛЬТАТЫ
print("\n" + "="*50)
print("РЕЗУЛЬТАТЫ ПРЕДОБРАБОТКИ")
print("="*50)

print(f"Исходная форма: {df.shape}")
print(f"После обработки: {df_processed.shape}")
print(f"Пропуски до: {df.isnull().sum().sum()}")
print(f"Пропуски после: {df_processed.isnull().sum().sum()}")
print(f"Типы данных после обработки:")
print(df_processed.dtypes.value_counts())

print("\nСтатистика после обработки:")
print(df_processed.describe())

# Сохранение результатов
df_processed.to_csv('temperature_preprocessed.csv', index=False)
print("\nОбработанные данные сохранены в 'temperature_preprocessed.csv'")
```

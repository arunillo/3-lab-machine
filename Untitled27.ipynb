```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

class DataPreprocessingPipeline:
    def __init__(self):
        self.scaler = None
        self.imputer = None

    def load_data(self):
        """Загрузка данных"""
        url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
        return pd.read_csv(url)

    def introduce_errors(self, df):
        """Добавление искусственных ошибок"""
        df_with_errors = df.copy()

        # 1. Пропущенные значения
        np.random.seed(42)
        missing_indices = np.random.choice(len(df), size=50, replace=False)
        df_with_errors.loc[missing_indices, 'Temp'] = np.nan

        # 2. Текстовые значения (сначала конвертируем в объект)
        df_with_errors['Temp'] = df_with_errors['Temp'].astype(object)
        text_indices = np.random.choice(len(df), size=10, replace=False)
        df_with_errors.loc[text_indices, 'Temp'] = 'error_value'

        # 3. Выбросы (только для числовых значений)
        numeric_values = pd.to_numeric(df_with_errors['Temp'], errors='coerce')
        outlier_indices = np.random.choice(numeric_values.dropna().index, size=15, replace=False)
        mean_val = numeric_values.mean()
        std_val = numeric_values.std()
        df_with_errors.loc[outlier_indices, 'Temp'] = mean_val + 5 * std_val

        return df_with_errors

    def visualize_data(self, df, title):
        """Визуализация данных"""
        fig, axes = plt.subplots(2, 2, figsize=(12, 8))

        # Только числовые данные для графиков
        numeric_data = pd.to_numeric(df['Temp'], errors='coerce')

        axes[0, 0].hist(numeric_data.dropna(), bins=30, alpha=0.7, color='skyblue')
        axes[0, 0].set_title(f'{title} - Распределение')

        axes[0, 1].boxplot(numeric_data.dropna())
        axes[0, 1].set_title(f'{title} - Box plot')

        # Временной ряд (первые 100 точек для наглядности)
        if 'Date' in df.columns:
            df_temp = df.head(100).copy()
            df_temp['Date'] = pd.to_datetime(df_temp['Date'])
            df_temp['Temp_numeric'] = pd.to_numeric(df_temp['Temp'], errors='coerce')
            axes[1, 0].plot(df_temp['Date'], df_temp['Temp_numeric'])
            axes[1, 0].set_title(f'{title} - Временной ряд')
            axes[1, 0].tick_params(axis='x', rotation=45)

        # Пропущенные значения
        missing = df.isna().sum()
        axes[1, 1].bar(missing.index, missing.values, color='coral')
        axes[1, 1].set_title(f'{title} - Пропуски')
        axes[1, 1].tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.show()

    def convert_to_numeric(self, df):
        """Конвертация в числовые типы"""
        df_processed = df.copy()

        print("Типы ДО конвертации:")
        print(df_processed.dtypes)

        # Конвертируем температуру
        df_processed['Temp'] = pd.to_numeric(df_processed['Temp'], errors='coerce')

        print("Типы ПОСЛЕ конвертации:")
        print(df_processed.dtypes)
        print(f"Не удалось конвертировать: {df_processed['Temp'].isna().sum() - df['Temp'].isna().sum()} значений")

        return df_processed

    def handle_missing_values(self, df):
        """Обработка пропусков"""
        missing_before = df.isna().sum()
        print(f"Пропуски до обработки: {missing_before['Temp']}")

        if missing_before['Temp'] > 0:
            self.imputer = SimpleImputer(strategy='mean')
            df_processed = df.copy()
            df_processed[['Temp']] = self.imputer.fit_transform(df_processed[['Temp']])
            print(f"Пропуски после обработки: {df_processed.isna().sum()['Temp']}")
        else:
            df_processed = df.copy()
            print("Пропусков нет - шаг пропущен")

        return df_processed

    def normalize_data(self, df):
        """Нормализация данных"""
        print("До нормализации:")
        print(df[['Temp']].describe())

        self.scaler = StandardScaler()
        df_processed = df.copy()
        df_processed[['Temp']] = self.scaler.fit_transform(df_processed[['Temp']])

        print("После нормализации:")
        print(df_processed[['Temp']].describe())

        return df_processed

    def run_pipeline(self):
        """Запуск пайплайна"""
        print("=== ЗАПУСК ПАЙПЛАЙНА ===")

        # 1. Загрузка
        df = self.load_data()
        print("Исходные данные:")
        print(df.head())

        # 2. Добавление ошибок
        df_with_errors = self.introduce_errors(df)
        print("\nПосле добавления ошибок:")
        print(f"Пропуски: {df_with_errors['Temp'].isna().sum()}")

        # 3. Визуализация с ошибками
        self.visualize_data(df_with_errors, "Данные с ошибками")

        # 4. Конвертация в числовые типы
        df_numeric = self.convert_to_numeric(df_with_errors)

        # 5. Обработка пропусков
        df_no_missing = self.handle_missing_values(df_numeric)

        # 6. Нормализация
        df_normalized = self.normalize_data(df_no_missing)

        # 7. Финальная визуализация
        self.visualize_data(df_normalized, "Обработанные данные")

        print("\n=== ПАЙПЛАЙН ЗАВЕРШЕН ===")
        print(f"Итоговый размер: {df_normalized.shape}")
        return df_normalized

# Запуск
pipeline = DataPreprocessingPipeline()
processed_data = pipeline.run_pipeline()
```
